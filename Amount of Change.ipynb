```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../MLB/csv/2020 Trevor Bauer.csv')
```

```python
plt.figure(figsize=(10, 10))
plt.xlim(-60, 60)
plt.ylim(-60, 60)

s = set()
for index, row in df.iterrows():
    x = row['pfx_x'] * (-30.48)
    z = row['pfx_z'] * 30.48
    if row['pitch_name'] == '4-Seam Fastball':
        if row['pitch_name'] not in s:
            s.add(row['pitch_name'])
            plt.scatter(x, z, color='red', label=row['pitch_name'])
        else:
            plt.scatter(x, z, color='red')
    if row['pitch_name'] == '2-Seam Fastball':
        if row['pitch_name'] not in s:
            s.add(row['pitch_name'])
            plt.scatter(x, z, color='brown', label=row['pitch_name'])
        else:
            plt.scatter(x, z, color='brown')
    elif row['pitch_name'] == 'Cutter':
        if row['pitch_name'] not in s:
            s.add(row['pitch_name'])
            plt.scatter(x, z, color='orange', label=row['pitch_name'])
        else:
            plt.scatter(x, z, color='orange')
    elif row['pitch_name'] == 'Slider':
        if row['pitch_name'] not in s:
            s.add(row['pitch_name'])
            plt.scatter(x, z, color='magenta', label=row['pitch_name'])
        else:
            plt.scatter(x, z, color='magenta')
    elif row['pitch_name'] == 'Curveball':
        if row['pitch_name'] not in s:
            s.add(row['pitch_name'])
            plt.scatter(x, z, color='pink', label=row['pitch_name'])
        else:
            plt.scatter(x, z, color='pink')
    elif row['pitch_name'] == 'Knuckle Curve':
        if row['pitch_name'] not in s:
            s.add(row['pitch_name'])
            plt.scatter(x, z, color='purple', label=row['pitch_name'])
        else:
            plt.scatter(x, z, color='purple')
    elif row['pitch_name'] == 'Split-Finger':
        if row['pitch_name'] not in s:
            s.add(row['pitch_name'])
            plt.scatter(x, z, color='blue', label=row['pitch_name'])
        else:
            plt.scatter(x, z, color='blue')
    elif row['pitch_name'] == 'Sinker':
        if '2-Seam Fastball' not in s:
            s.add('2-Seam Fastball')
            plt.scatter(x, z, color='brown', label='2-Seam Fastball')
        else:
            plt.scatter(x, z, color='brown')
    elif row['pitch_name'] == 'Changeup':
        if row['pitch_name'] not in s:
            s.add(row['pitch_name'])
            plt.scatter(x, z, color='lime', label=row['pitch_name'])
        else:
            plt.scatter(x, z, color='lime')

plt.legend(loc="upper left", fontsize=14)
plt.show()
```
